# Areawise data 

```python
# Column name
df.columns
```

```python
swiggy_area = df.groupby(['Area'], as_index= False)['Total ratings'].sum().sort_values(by='Total ratings',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
ax=sns.barplot(x='Area',y= 'Total ratings',data=swiggy_area,palette = 'Set3')
for bars in ax.containers:
    ax.bar_label(bars)
```

This graph shows us top 10 areas that most of the customer gives the good overall ratings.From above graphs we can see that 
most of the customers are in chembur,banjara hills, rohini.

```python
swiggy_area = df.groupby(['Area'], as_index= False)['Avg ratings'].sum().sort_values(by='Avg ratings',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
ax=sns.barplot(x='Area',y= 'Avg ratings',data=swiggy_area,palette = 'Set3')
for bars in ax.containers:
    ax.bar_label(bars)
```

Here we can see clearly as total rating and average rating ,here rohini,chembur,and andheri east those areas are come in top 3 position

```python
ax = sns.countplot(x = 'City',data = df,palette = 'pastel')
sns.set(rc={'figure.figsize':(10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
swiggy_city = df.groupby(['City'], as_index= False)['Avg ratings'].sum().sort_values(by='Avg ratings',ascending=False).head(13)
sns.set(rc={'figure.figsize':(13,5)})
ax=sns.barplot(x='City',y= 'Avg ratings',data=swiggy_city,palette = 'Set3')
for bars in ax.containers:
    ax.bar_label(bars)
```

from above graph we can see that those city who has great average rating.

```python
swiggy_city = df.groupby(['City'], as_index= False)['Total ratings'].sum().sort_values(by='Total ratings',ascending=False).head(13)
sns.set(rc={'figure.figsize':(13,5)})
ax=sns.barplot(x='City',y= 'Total ratings',data=swiggy_city,palette ='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
```

Again same here from above graph we can see that those city who has great total rating.

```python
swiggy_city = df.groupby(['City'],as_index=False)['Delivery time'].sum().sort_values(by='Delivery time',ascending=False).head(13)
sns.set(rc={'figure.figsize':(13,5)})
ax = sns.barplot(x='City',y= 'Delivery time',data=swiggy_city,palette = 'Set2')
for bars in ax.containers:
    ax.bar_label(bars)
```

As we can see here delivery time according their city.

# Food

```python
swiggy_food=df.groupby(['Food type'],as_index= False)['Total ratings'].sum().sort_values(by='Total ratings',ascending=False).head(6)
sns.set(rc={'figure.figsize':(13,5)})
ax=sns.barplot(x='Food type',y= 'Total ratings',data=swiggy_food,palette = 'muted')
for bars in ax.containers:
    ax.bar_label(bars)
```

Above graph we can see here top 6 food types the customer gives the great rating here.

```python
swiggy_food=df.groupby(['Food type'],as_index= False)['Avg ratings'].sum().sort_values(by='Avg ratings',ascending=False).head(6)
sns.set(rc={'figure.figsize':(13,5)})
ax=sns.barplot(x='Food type',y= 'Avg ratings',data=swiggy_food,palette = 'muted')
for bars in ax.containers:
    ax.bar_label(bars)
```

Above graph we can see here top 6 food types the customer gives the good average rating here.

```python
swiggy_food=df.groupby(['Food type'],as_index= False)['Price'].sum().sort_values(by='Price',ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='Food type',y= 'Price',data=swiggy_food,palette = 'rocket')
for bars in ax.containers:
    ax.bar_label(bars)
```

Here we can see indian,chinese,and north indian food has a high price

```python
swiggy_food=df.groupby(['Food type'],as_index= False)['Delivery time'].sum().sort_values(by='Delivery time',ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
ax=sns.barplot(x='Food type',y= 'Delivery time',data=swiggy_food,palette = 'rocket')
for bars in ax.containers:
    ax.bar_label(bars)
```

Here we clearly see that indian,chinese,and north indian food take more time to preapre as compared to other food.
